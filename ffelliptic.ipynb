```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plot the k-rational points of an elliptic curve

def plotfiniteelliptic(a,b,p):

    points = []
    for x in range(p):
        for y in range(p):
            if (y**2) % p == (x**3 + a*x + b) % p:
                points.append([x,y])
                
    # plot the points
    points = np.array(points)
    plt.scatter(points[:,0],points[:,1])
    plt.title(r'k-rational points of the elliptic curve $y^2 = x^3 + {}x + {}$ over $F$'.format(a,b) + str(p))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    
    return len(points)


p = 19
a = -1
b= 1


x = np.arange(p)
y = np.arange(p)
X, Y = np.meshgrid(x, y)
Z = (Y**2 - (X**3 + a*X + b)) % p

sns.set_style('white')
plt.figure(figsize=(8,8))
# make margins smaller
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)
plt.scatter(X[Z==0], Y[Z==0], s=100, c='k')


plt.title('K-rational Points of $y^2 = x^3 - x + 1$ over $F_{19}$', fontsize=20)


plt.xlim(-1, p)
plt.ylim(-1, p)
plt.xticks(np.arange(p))
plt.yticks(np.arange(p))
plt.grid()
# plt.show()

# plotfiniteelliptic(a,b,p)





# plt.show()
plt.savefig('krationalpoints.png', dpi=300)
```
