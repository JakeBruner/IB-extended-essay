```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plot the polylogarithm function for Li_2(z), Li_3(z), Li_4(z)
# for z in the range [0, 1]
# the polylogarithm function is defined as
# Li_n(z) = sum_{k=1}^inf z^k / k^n

def polylog(z, n):
    k = 1
    sum = 0.0
    while True:
        term = z**k / k**n
        if term < 1e-10:
            break
        sum += term
        k += 1
    return sum
  
```

```python
# plot the polylogarithm function for Li_2(z), Li_3(z), Li_4(z)
# for z in the range [0, 1]

z = np.linspace(0, 1, 100)
li2 = [polylog(zz, 2) for zz in z]
li3 = [polylog(zz, 3) for zz in z]
li4 = [polylog(zz, 4) for zz in z]

plt.plot(z, li2, label='Li_2(z)')
plt.plot(z, li3, label='Li_3(z)')
plt.plot(z, li4, label='Li_4(z)')
plt.xlabel('z')
plt.ylabel('Li_n(z)')
plt.legend()
plt.show()
```
